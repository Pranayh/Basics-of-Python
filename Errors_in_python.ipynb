# **Python Error Types:**


The most common reason of an error in a Python program is when a certain statement is not in accordance with the prescribed usage. Such an error is called a syntax error. The Python interpreter immediately reports it, usually along with the reason.


```
>>> print "hello"
```

**IndexError:** 
It is thrown when trying to access an item at an invalid index.

```
>>> L1=[1,2,3]
>>> L1[3]
```

**ModuleNotFoundError:** 
It is thrown when a module could not be found.

```
>>> import notamodule
```

**KeyError:** It is thrown when a key is not found.

```
>>> D1={'1':"aa", '2':"bb", '3':"cc"}
>>> D1['4']
```

**ImportError:** It is thrown when a specified function can not be found.

```
>>> from math import cube
```

**StopIteration:**It is thrown when the next() function goes beyond the iterator items.

```
>>> it=iter([1,2,3])
>>> next(it)

>>> next(it)

>>> next(it)

>>> next(it)
```

**TypeError:**It is thrown when an operation or function is applied to an object of an inappropriate type.

```
>>> '2'+2
```

**ValueError:**It is thrown when a function's argument is of an inappropriate type.

```
>>> int('xyz')
```

**NameError:**It is thrown when an object could not be found.

```
>>> age
```

**ZeroDivisionError:**It is thrown when the second operator in the division is zero.

```
>>> x=100/0
```

**KeyboardInterrupt:**It is thrown when the user hits the interrupt key (normally Control-C) during the execution of the program.

```
>>> name=input('enter your name')
```
